```python
import numpy as np
import matplotlib.image as mtimg
import matplotlib.pyplot as plt
import random
from skimage import transform
from skimage.util import random_noise
import cv2
```

```python
img = mtimg.imread('cat.jpg')
```

```python
# img size = 660 * 710
plt.imshow(img)
```

#### Rescaling

```python
image_rescale = cv2.resize(img, (150, 150))
```

```python
plt.imshow(image_rescale)
```

```python
image_rescale2 = cv2.resize(img, (50, 50))
plt.imshow(image_rescale2)
```

#### FLIPPING

```python
horizontal_flip = np.fliplr(img)
```

```python
plt.imshow(horizontal_flip)
```

```python
vertical_flip = np.flipud(img)
plt.imshow(vertical_flip)
```

#### Rotation

```python
trans_img = transform.rotate(img, random.uniform(-40,40))
plt.imshow(trans_img)
```

#### Adding Noise

```python
img_noise = random_noise(img, mode='s&p', clip=True)
plt.imshow(img_noise)
```

#### Image to Array

```python
gray_scale = cv2.imread('./cat.jpg', cv2.IMREAD_GRAYSCALE)
gray_scale.shape
```

```python
new_array = cv2.resize(gray_scale, (100,100))
new_array
```

```python
plt.imshow(new_array, cmap='gray')
plt.title("Resized Image")
plt.show()
```
